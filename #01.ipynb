必要なライブラリをインポート

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Numpy**は様々な数学関数を含んだライブラリであり、三角関数や行列といった高度な数学的操作やデータ分析を実行できる。

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

グリッドを追加

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

タイトル、軸ラベル追加

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.title('sin wave')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.cos(x)

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.cos(x)

plt.plot(x,y)
plt.title('cos wave')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

sinとcosのグラフを重ねてみる

```python
x = np.arange(0, 2*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')

plt.grid()
plt.title('sin & cos wave')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
x = np.random.randn(1000)

plt.hist(x)
plt.title('histgram')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

ヒストグラムデータが2つあるとき

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(1000)

bins = [ -3.5, -2.5, -1.5, -0.5, 0.5, 1.5, 2.5, 3.5]
plt.hist(x1, bins, label='x1')
plt.hist(x2, bins, label='x2', histtype='step')

plt.title('histgram')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

ランダムデータから区間が決定するため、区間が一致しないのでbinsであらかじめ区間を設定した。  
上書きされてx1のヒストグラムが隠れてしまうのを防ぐために、histtype='step'で塗りつぶし無しを指定。

ヒストグラムを並列で表示させる

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(1000)

labels = ['x1','x2']
x = np.hstack([np.c_[x1], np.c_[x2]])
plt.hist(x, label=labels)

plt.title('histgram')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

重ね書きする個々のデータ（x1,x2）を列方向に並べて連結した行列を作成し、それをhist()関数に与える。  

$$
    \boldsymbol{a} = (a_{1},a_{2},\dots,a_{n}),　　　\boldsymbol{b} = (b_{1},b_{2},\dots,b_{n})  
$$  

$$
    \boldsymbol{X} =
        \begin{pmatrix}
            a_{1} & b_{1} \\
            a_{2} & b_{2} \\
            \vdots  & \vdots \\
            a_{n} & b_{n} \\
        \end{pmatrix}
$$

行列の作成には、Numpyのc_関数やhstack()関数を使用できる。

積み上げ型ヒストグラム

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(1000)

labels = ['x1', 'x2']
x = np.hstack([np.c_[x1], np.c_[x2]])

plt.hist(x, label=labels, stacked=True)

plt.title('histgram')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

#### 等値線図

np.meshgrid関数を用いて格子列$(x,y)$を生成し、格子ごとに$z = \sqrt{x^2+y^2}$の値を求めてプロット

```python
x1 = np.linspace(-2,2,100)
y1 = np.linspace(-2,2,100)

x, y = np.meshgrid(x1,y1)
z = np.sqrt(x**2 + y**2)

plt.title('Equal Diagram')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.contour(x,y,z)
plt.show()
```

$z=sin(x^2+y^2)$をカラーでプロット

```python
x1 = np.linspace(-2,2,100)
y1 = np.linspace(-2,2,100)

x, y = np.meshgrid(x1,y1)
z = np.sin(x**2 + y**2)

plt.title('Color Equal Diagram')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.contourf(x,y,z, cmap='RdBu')
plt.colorbar()
plt.show()
```
